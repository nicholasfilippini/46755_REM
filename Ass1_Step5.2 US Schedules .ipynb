```python
import matplotlib.pyplot as plt
import numpy as np

Wind_production = [6159.269999999999, 6300.3099999999995, 3779.9, 3509.4700000000003, 3753.6, 2469.8000000000006]

Electrolizers_downward_reserve_total = [-1208.383333333333, -1240.4808333333328]
Electrolizers_upward_reserve_total = [0.0, 0.0]


width = 0.2 # narrower width for conventional_units_prod
indices = np.arange(len(Wind_production)) + 1 # shift x-axis ticks by 1
indices1 = np.arange(len(Electrolizers_downward_reserve_total)) +1


plt.bar(indices, Wind_production, width=0.3, 
        color='lightsteelblue', label='Production')
plt.bar([i+ width for i in indices1], Electrolizers_downward_reserve_total, 
        width=0.3, color='mediumaquamarine', alpha=0.8, label='Downward Reserve')
plt.bar([i+ width for i in indices1], Electrolizers_upward_reserve_total, 
        width=0.3, color='gold', alpha=0.8, label='Upward Reserve')

plt.legend(loc='upper right', title='Wind Power Units US Schedule', fontsize='small')

plt.xticks(indices)
plt.xlabel('Units', fontweight='bold')
plt.ylabel('Power (MWh)', fontweight='bold')
plt.grid()
plt.show()
```
